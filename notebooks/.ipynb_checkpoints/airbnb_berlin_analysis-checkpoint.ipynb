
# Airbnb Listings Analysis in Berlin

This project analyzes Airbnb listing data in Berlin to understand pricing patterns
and factors that influence nightly prices. The analysis focuses on room type,
location, and demand indicators such as the number of reviews.

The dataset used is publicly available and anonymized.


```python

import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python

# Load raw data (raw data is never modified or overwritten)
df = pd.read_csv("../data/raw/listings.csv")
```

```python

# Create figures directory if it does not exist
os.makedirs("../figures", exist_ok=True)
```


## Data Cleaning

The price column contains missing values. Since price is central to the analysis,
missing values are imputed using the median price for each room type.


```python

df['price'] = df.groupby('room_type')['price'].transform(
    lambda x: x.fillna(x.median())
)
```


To focus on typical market behavior, listings with prices outside the range
of 20 to 500 per night are removed.


```python

df = df[df['price'].between(20, 500)]
```


For the analysis of price versus reviews, listings with more than 300 reviews
are excluded to improve interpretability.


```python

df_reviews = df[df['number_of_reviews'] <= 300]
```


## Exploratory Data Analysis
### Price Distribution


```python

plt.hist(df['price'], bins=30)
plt.xlabel('Price per night')
plt.ylabel('Number of listings')
plt.title('Distribution of Airbnb Prices in Berlin')
plt.savefig("../figures/price_distribution.png", dpi=300, bbox_inches="tight")
plt.show()
```


### Price by Neighbourhood

The analysis focuses on the top 10 neighbourhoods by number of listings to ensure
meaningful comparisons.


```python

top_neighbourhoods = df['neighbourhood'].value_counts().head(10).index
df_top = df[df['neighbourhood'].isin(top_neighbourhoods)]

median_prices = (
    df_top.groupby('neighbourhood')['price']
    .median()
    .sort_values(ascending=False)
)

median_prices.plot(kind='bar')
plt.ylabel('Median price per night')
plt.title('Median Airbnb Price by Neighbourhood (Top 10)')
plt.xticks(rotation=45, ha='right')
plt.savefig("../figures/price_by_neighbourhood.png", dpi=300, bbox_inches="tight")
plt.show()
```


### Price by Room Type


```python

median_room_prices = (
    df.groupby('room_type')['price']
    .median()
    .sort_values(ascending=False)
)

median_room_prices.plot(kind='bar')
plt.ylabel('Median price per night')
plt.title('Median Airbnb Price by Room Type')
plt.savefig("../figures/price_by_room_type.png", dpi=300, bbox_inches="tight")
plt.show()
```


### Price vs Number of Reviews


```python

plt.scatter(
    df_reviews['number_of_reviews'],
    df_reviews['price'],
    alpha=0.3
)
plt.xlabel('Number of reviews')
plt.ylabel('Price per night')
plt.title('Price vs Number of Reviews')
plt.savefig("../figures/price_vs_reviews.png", dpi=300, bbox_inches="tight")
plt.show()
```


## Key Findings

- Prices in Berlin are concentrated in the mid-range
- Room type is a stronger driver of price than neighbourhood
- Location affects price, but differences between popular neighbourhoods are moderate
- Higher prices do not necessarily lead to higher demand
